```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


t = np.linspace(0, 2 * np.pi, 100)
x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 1 * np.cos(2* t) - 2 * np.cos(2 * t) - np.cos(2 * t)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot(x, y, t, color='red')


ax.set_xlim(-20, 20)
ax.set_ylim(-20, 20)
ax.set_zlim(0, 2 * np.pi)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('T')


plt.show()
```
